### outliers

```python
Q1= data.price.quantile(.25)
Q3= data.price.quantile(.75)
Q1,Q3
```

### finding iqr

```python
IQR= Q3-Q1
IQR
```

### CALCULATE IQR

```python
lower_limit= Q1 - 1.5 * (IQR)
upper_limit= Q3 + 1.5 * (IQR)
lower_limit,upper_limit
```

### outliers

```python
data[(data.price<lower_limit)|(data.price>upper_limit)]
```

### data with no outliers

```python
data_no_outliers = data[(data.price>lower_limit) & (data.price< upper_limit)]
```
